### Notes 
- No duplicates entries
- all entries seem to be the correct type
  - except for EXCEEDING MINIMUM INCOME RESTRICTIONS (20 PTS)             91 non-null     float64
  - which should be changed to int64
- fix all NaN values by filling in with appropriate data, no more missing values in the table

```python
df.info()
```

```python
col = 'EXCEEDING MINIMUM INCOME RESTRICTIONS (20 PTS)'
if (df[col] % 1 != 0).any():
    print(f"Warning: {col} has non-integer values; rounding down")
    df[col] = df[col].round(0)
# Convert to int64
df[col] = df[col].astype("int64")
```

### Requested Amounts
- Recall:
  - Average *BOND REQUEST* is `$37.6M`, average *ANNUAL FEDERAL CREDIT REQUEST* is `$3.34M`
  - Most projects have zero *STATE CREDIT REQUEST* (except for one), shows a reliance on federal credit rather than state credit

### Homeless Units
- Recall:
  - Most projects (78%) have close to no homeless units, but some (5.5%) dedicate all of their units to homeless
  - 

```python
plt.figure(figsize=(4, 2))  # Set the size of the figure
sb.histplot(data = df, x = "UNITS FOR HOMELESS", stat = "percent", bins = 20)
```

```python
plt.figure(figsize=(8, 2))  # Set the size of the figure
ax = sb.histplot(data = df, x = "HOMELESS %", stat = "percent", bins = 20)
for patch in ax.patches:
    height = patch.get_height()
    if height > 0:  # Only annotate non-zero bars
        ax.annotate(
            f'{height:.1f}%', 
            (patch.get_x() + patch.get_width() / 2, height), 
            ha='center', va='bottom', fontsize=10
        )
```
